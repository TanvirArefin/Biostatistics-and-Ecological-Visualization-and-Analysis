# Descriptive Statistics for numeric columns

```python
#Descriptive Statistics(Mean,Median,Min, Max)
columns= ['Temperature (°C)', 'Pollution_Level (ppm)', 'Species_Population']

def stats(data, columns):
    stats = {}
    for col in columns:
        stats[col] = {
            'mean': data[col].mean(),
            'median': data[col].median(),
            'min': data[col].min(),
            'max': data[col].max()
        }
    return stats

results = stats(data, columns)

# Print the results
for col, stats in results.items():
    print(f"Statistics for {col}:")
    for stat_name, stat_value in stats.items():
        print(f"  {stat_name}: {stat_value}")
```

*   The average temperature is 22.7°C, with a median of 23°C. The temperatures range from 19.7°C to 25.5°C.
*   The average pollution level is 14.2 ppm, with a median of 14 ppm. The levels range from 10 ppm to 18 ppm.
*  The average species population is 1786.9, with a median of 1725. The population ranges from 800 to 3100.

# Multiple Linear Regression Model

```python
import statsmodels.api as sm
X = data[['Temperature (°C)', 'Species_Population']]  
X = sm.add_constant(X)
y = data['Pollution_Level (ppm)'] 

# Fit the regression model
model = sm.OLS(y, X).fit()
print(model.summary())

# Scatter plot
sns.lmplot(x='Temperature (°C)', y='Pollution_Level (ppm)', data=data, aspect=1.5, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
sns.lmplot(x='Species_Population', y='Pollution_Level (ppm)', data=data, aspect=1.5, scatter_kws={'color': 'green'}, line_kws={'color': 'orange'})
plt.show()
```

**Model Explanation**

1.   The model explains 78.6% of the variation in pollution levels, meaning it fits the data well overall.
2.   The condition number (4.28e+04) indicates potential multicollinearity issues, meaning the predictors might be too closely related, affecting accuracy.
3.   Species Population significantly increases pollution levels (p = 0.001), while Temperature does not show a strong effect (p = 0.421).
4.   No strong evidence of outliers or non-normality in the residuals was detected (Jarque-Bera test p = 0.668).
5.   The intercept of 6.37 suggests pollution starts at 6.37 when predictors are zero, but it's not very reliable.

# User Interface

```python
#User Interface

import ipywidgets as widgets
from IPython.display import display, clear_output

# Function to add a new row to the DataFrame
def add_data(temperature, pollution, population, region, timestamp):
    new_row = pd.DataFrame({'Temperature (°C)': [temperature],
                            'Pollution_Level (ppm)': [pollution],
                            'Species_Population': [population],
                            'Region': [region],
                            'Timestamp': [timestamp]})
    return pd.concat([data, new_row], ignore_index=True)

# Function to delete a row from the DataFrame
def delete_data(index):
    global data 
    if 0 <= index < len(data):
        data = data.drop(index)
        data = data.reset_index(drop=True)
        print(f"Row {index} deleted successfully.")
    else:
        print("Invalid index. No row deleted.")

# Widgets for input
temperature_input = widgets.FloatText(description='Temperature:')
pollution_input = widgets.FloatText(description='Pollution:')
population_input = widgets.IntText(description='Population:')
region_input = widgets.Dropdown(options=['Zone 1', 'Zone 2', 'Zone 3'], description='Region:')
timestamp_input = widgets.Text(description='Timestamp:')

add_button = widgets.Button(description='Add Data')
delete_index_input = widgets.IntText(description='Index to delete:')
delete_button = widgets.Button(description='Delete Data')

output = widgets.Output()


def on_add_button_clicked(b):
  with output:
    clear_output()
    global data
    data = add_data(temperature_input.value, pollution_input.value, population_input.value, region_input.value, timestamp_input.value)
    print("Data added successfully!")
    print(data.tail())  
def on_delete_button_clicked(b):
  with output:
    clear_output()
    delete_data(delete_index_input.value)
    print(data.tail())


add_button.on_click(on_add_button_clicked)
delete_button.on_click(on_delete_button_clicked)


display(temperature_input, pollution_input, population_input, region_input, timestamp_input, add_button, delete_index_input, delete_button, output)
```
